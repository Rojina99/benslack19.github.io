```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ggplot import *
```

```python
df = pd.DataFrame()   # initialize an empty dataframe\n",
df['feature_x'] = np.arange(-10, 10)             # to be plotted in x-dimension\n",
df['feature_y'] = np.arange(-20, 20, 2)
df['feature_z'] = np.random.normal(size=20) # to be plotted in y-dimension\n",
groupSize = round(len(df)/5)
df['feature_cat'] = ['group 1']*groupSize + ['group 2']*groupSize + ['group 3']*groupSize + ['group 4']*groupSize + ['group 5']*(len(df)-groupSize*4)
```

```python
# however the bottom point drops out
ggplot(aes(x='feature_x', y='feature_y', color='feature_cat', size='feature_y'), data=df) + geom_point()
```
